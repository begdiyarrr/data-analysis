```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path = 'dataframes/taxi_peru.csv'
calls = pd.read_csv(path, sep=';')
```

```python
num_calls = calls.source.value_counts().sum()
max_calls = calls.source.value_counts().max()
print(round(max_calls/num_calls*100))
```

```python
sns.barplot(calls['source'].value_counts(), palette='Set1')
```

```python
# Создаем график
plt.figure(figsize=(6, 4))
sns.countplot(data=calls, x="source", hue="end_state")

# Добавляем подписи осей и заголовок
plt.xlabel("Платформа")
plt.ylabel("Количество")
plt.title("Распределение итогового состояния заказа по платформам")

# Отображаем график
plt.show()
```

```python
driver_score_counts = round(calls.driver_score.value_counts() / calls.driver_score.count() * 100, 2).reset_index()
driver_score_counts = driver_score_counts.rename(columns={'index':'driver_score', 'count':'percentage'}).sort_values('driver_score')
print(driver_score_counts)
```

```python
ax = sns.barplot(x='driver_score', y='percentage', data=driver_score_counts, color='blue', alpha=0.5)
ax.set(xlabel='Driver score', ylabel='Percentage')
sns.despine()  # убрать часть рамки графика
plt.show()
```

```python
rider_score_counts = round(calls.rider_score.value_counts() / calls.rider_score.count() * 100, 2).reset_index()
rider_score_counts = rider_score_counts.rename(columns={'index':'rider_score', 'count':'percentage'}).sort_values('rider_score')
print(rider_score_counts)
```

```python
rx = sns.barplot(x='rider_score', y='percentage', data=rider_score_counts, color='red', alpha=0.5)
rx.set(xlabel='Rider score', ylabel='Percentage')
sns.despine()
plt.show()
```
